#### Import Libraries and load in dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load only the data within the columns
df = pd.read_excel("Health Insurance Dataset.xlsx", nrows=1338)
```

## Univariate Statistics - Continous Variables

#### Age

```python
# Plot histograms
sns.histplot(df['age']).set(title='Age Distribution')
plt.show()
```

#### Charges

```python
sns.histplot(df['charges']).set(title='Charges Distribution')
plt.show()
```

## Univariate Statistics - Categorical Variables

#### Sex

```python
# Plot bar charts
sns.countplot(x='sex', data=df).set(title='Sex Distribution')
plt.show()
```

#### Smoker

```python
sns.countplot(x='smoker', data=df).set(title='Smoker Distribution')
plt.show()
```

## Bivariate Statistics - Continuous Variables
#### BMI vs Health Score

```python
sns.regplot(x='bmi', y='score', data=df, line_kws={"color": "red"}).set(title='BMI vs Health Score')
plt.show()
```

## Bivariate Statistics - Categorical Variables
#### Sex vs Smoker

```python
sns.countplot(x='sex', hue='smoker', data=df).set(title='Sex vs Smoker')
plt.show()
```
